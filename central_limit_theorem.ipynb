```python
from scipy.stats import beta
import altair as alt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
a, b = (2,10)
```

```python
x = np.linspace(0,1,500)
y = [beta.pdf(x, a, b) for x in x]
```

```python
df = pd.DataFrame()

df['x'] = x
df['y'] = y
```

```python
alt.Chart(df).mark_line().encode(
    x='x',
    y='y'
)
```

```python
for s in [2,3,4,5,6,7,8,9,10]:
    vals = [beta.rvs(2, 10, size=s).mean() for x in range(1000000)]
    sns.kdeplot(vals, shade=False)
```
